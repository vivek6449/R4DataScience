# Position Adjustments

```r
ggplot(data = diamonds) + geom_bar(mapping = aes(x = cut, color = cut))
```

```r
ggplot(data = diamonds) + geom_bar(mapping = aes(x = cut, fill = cut))
```

```r
# If we map the fill aesthetic to another variable like "clarity"
ggplot(data = diamonds) + geom_bar(mapping = aes(x = cut, fill = clarity))
```

```r
# If we don't want position argument we have another three options. 1) identity 2) dodge 3) fill
ggplot(data = diamonds, mapping = aes(x= cut, fill = clarity)) + geom_bar(alpha = 1/5, position = "identity")
# Here alpha will decide intensity of color
```

```r
ggplot(data = diamonds, mapping = aes(x = cut, color = clarity)) + geom_bar(fill = NA, position = "identity")
```

```r
ggplot(data = diamonds) + geom_bar(mapping = aes(x = cut, fill = clarity), position = "fill")
```

```r
# This places overlapping objects directly one another
ggplot(data = diamonds) + geom_bar(mapping = aes(x = cut, fill = clarity), positon = "dodge")
```

# The layered grammer of graphics (FORMAT)

```r
ggplot(data = <DATA>)+<GEOM_FUNCTION>(mapping = aes(<MAPPINGS>)), stat = <STAT>, position = <POSITION>)
+ <COORDINATE_FUNCTION> + <FACET_FUNCTION>
```

```r
ggplot(nz, aes(long, lat, group = group)) + geom_polygon(fill = "white", color = "black")
```

```r
# coord_map = projects a portion of the earth, which is approximately spherical, onto a flat 2D plane using any projection
# preserve straight lines, so this requires considerable computation. 
# coord_quickmap = quick approximation that does preserve straight lines
ggplot(nz, aes(long, lat, group = group)) + geom_polygon(fill = "white", color = "black")+coord_quickmap()
```

# Logical Operators

```r
# Don't confuse with filter(flights, month == 11 | 12)
filter(flights, month == 11 | month == 12)
```

```r
# Can also write -- " x %in%y "
nov_dec  <- filter(flights, month %in% c(11,12))
```

```r
# Can also apply De Morgan's law
# !(x & y) = !x | !y and !(x | y) = !x & !y
# if we want to see flights that weren't delayed(on arrival or departure) by more than two hours
filter(flights, !(arr_delay > 120 | dep_delay > 120))
```

```r
# same as above
filter(flights, arr_delay <= 120 , dep_delay <= 120)
```

# Add new variables with "mutate()"

```r
# mutate() = Besides selecting sets of existing columns, it’s often useful to add new columns that are functions of existing columns.
# We will create small dataset 
flights_sml  <- select (flights, year:day, ends_with("delay"), distance, air_time)
# ends_with("delay") will include all the terms which include delay
```

```r
mutate(flights_sml, gain = arr_delay - dep_delay, speed = distance / air_time * 60)
```

```r
# We can refer to columns that we have just created
mutate(flights_sml, gain = arr_delay - dep_delay, hours = air_time/60, gain_per_hour = gain/hours)
```

```r
# If we want to keep the new variables ONLY
transmute(flights, gain=arr_delay - dep_delay, hours = air_time/60, gain_per_hour = gain/hours)
```

```r
# This will convert dep_time into hour and minute
transmute(flights, dep_time, hour = dep_time %/% 100, minute = dep_time%%100)
```

# Useful summary functions

```r
# Median = median is a value where 50% of x is above it and 50% is below it
not_cancelled %>%
group_by(year, month, day) %>%
summarise(avg_delay1 = mean(arr_delay), avg_delay2 = mean(arr_delay[arr_delay > 0]))
```

```r
# Measures of spread: sd(x) = standard deviation, IQR = Interquartile range, mad = median absolute deviation
not_cancelled %>%
group_by(dest) %>%
summarise(distance_sd = sd(distance)) %>%
arrange(desc(distance_sd))
```

```r
# when do the first and last flights leave each day?
not_cancelled %>%
group_by(year, month, day) %>%
summarise(first = min(dep_time), last = max(dep_time))
```

```r
# Measure of position: first(), nth(x,2), last(x)
not_cancelled %>%
group_by(year, month, day) %>%
summarise(first_dep = first(dep_time), last_dep = last(dep_time))
```

```r
# This will give you first and last dep. time in each row 
not_cancelled %>%
group_by(year,month,day) %>%
mutate(r = min_rank(desc(dep_time))) %>%
filter (r %in% range(r))
```

```r
# Counts: we've seen n(), which take no arguments and returns the size of the current group
# To count the no. of non-missing values=sum(!is.na(x)), for no. of distinct values=n_distince(x)
not_cancelled %>%
group_by(dest) %>%
summarise(carriers = n_distinct(carrier)) %>%
arrange(desc(carriers))
```

```r
# Sometimes counts are so useful
not_cancelled %>%
count(dest)
```

```r
# Could use this to count the total number of miles a plane flew
# We can optionally provide a weight variable
not_cancelled %>%
count(tailnum, wt=distance)
```

```r
# How many flights left before 5am? (these usually indicate delayed flights from the previous day)
# This sum will give total number of delayed flights
not_cancelled %>%
group_by(year, month, day) %>%
summarise(n_early = sum(dep_time < 500))
```

```r
# USE OF "mean" = what propotion of flights are delayed by more than hour?
not_cancelled %>%
group_by(year,month, day) %>%
summarise(hour_perc=mean(arr_delay > 60))
```

# Grouped mutates (and filters)

```r
# Grouping is most useful in conjunction with summarise(), but you can also do convenient operations with mutate() and filter()
# Find the worst member of each group ( means for each day (lowest 10))
flights_sml %>%
group_by(year, month, day) %>%
filter(rank(desc(arr_delay)) < 10)
```

```r
# Find all the groups bigger than a threshold
popular_dest  <- flights %>%
group_by(dest) %>%
filter(n() > 365)
popular_dest
```

```r
# Standardise to compute per group metrics
popular_dest %>%
filter(arr_delay > 0) %>%
mutate(prop_delay = arr_delay / sum(arr_delay)) %>%
select(year:day, dest, arr_delay, prop_delay)
```

```r
# which plane(tailnum) has the worst on-time record
flights %>%
  group_by(tailnum) %>%
  summarise(mean_arr_delay = mean(arr_delay, na.rm=TRUE),
            flights = n()) %>%
  arrange(desc(mean_arr_delay))
```

```r
# What time of day should you fly if you want to avoid delays as much as possible?
flights %>%
    ggplot(aes(x=factor(hour), fill=arr_delay > 5 | is.na(arr_delay))) + geom_bar()
```

```r
flights %>%
group_by(hour) %>%
summarise(hour_arr_delay = sum(arr_delay , na.rm = TRUE)/n(),flights = n()) %>%
arrange(hour_arr_delay)
```

```r
# For each destination, compute the total minutes of delay.
flights %>%
group_by(dest) %>%
summarise(total_delay = sum(arr_delay + dep_delay, na.rm = TRUE))
arrange(total_delay)
```

```r
# For each, flight, compute the proportion of the total delay for its destination.
not_cancelled %>%
group_by(tailnum) %>%
summarise(prop_delay = mean((arr_delay + dep_delay)/n()))
arrange(prop_delay)
```

```r
# Delays are typically temporally correlated: even once the problem that caused the initial delay 
# has been resolved, later flights are delayed to allow earlier flights to leave. Using lag() explore 
# how the delay of a flight is related to the delay of the immediately preceding flight.
flights %>%
mutate(new_dep_time = lubridate::make_datetime(year, month, day, hour, minute)) %>%
group_by(origin) %>%
arrange(new_dep_time) %>%
mutate (prev_dep_delay = lag(dep_delay)) %>%
arrange(new_dep_time)
```
